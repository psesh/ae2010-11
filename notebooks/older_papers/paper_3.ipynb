### Problem 2

Consider a control volume of a set of streamlines describing the flow up to the noise of an airfoil. The streamlines are stationary in space, so there is no external work done on the fluid. Additionally assume there is no heat transferred to the flow. 

\begin{enumerate}
\item[a.] Consider two points along a single streamline: one in the freestream and one at the nose of the airfoil. From the steady flow momentum equation, please describe the static temperature variation along this streamline. ${\color[rgb]{0.028509,0.250925,0.501969}[4]}$

If we consider a control volume as shown above, the steady flow energy equation takes the form

h_2 + \frac{v_2^2}{2} & = h_1 + \frac{v_1^2}{2} \\ 
\Rightarrow c_p T_2 + \frac{v_2^2}{2} & = c_p T_1 + \frac{v_1^2}{2} \\ 
\Rightarrow T_2 + \frac{v_2^2}{2c_p} & = T_1 + \frac{v_1^2}{2c_p} 
\end{split}

\item[b.] From the above, please write down the equation for the stagnation temperature, and describe it in words, e.g., ``It is the temperature that the fluid would reach...''. ${\color[rgb]{0.028509,0.250925,0.501969}[2]}$

Despite the above error, students are expected to have been familiar with Slides 17-18 in Lecture 21: the definition of stagnation temperature. 

Now setting station $2$ to be the stagnation one, we have

\begin{equation}
T_{01}  = T_1 + \frac{v_1^2}{2 c_p} 
\end{equation}

It is the temperature that the fluid would reach if it were brought to a velocity of zero by a steady, isentropic process with no external work. 

}
\item[c.] From b, please derive the following expression ${\color[rgb]{0.028509,0.250925,0.501969}[4]}$:

\begin{equation*}
\frac{{\color[rgb]{0.164799,0.878862,0.723179}T_{01}}}{{\color[rgb]{0.164799,0.878862,0.723179}T_1}} = 1 + \frac{\left(\gamma -1 \right)}{2}{\color[rgb]{0.041732,0.352132,0.699576}M_1}^2
\end{equation*}


{\color{RubineRed} 

From 

\begin{equation*}
T_{01}  = T_1 + \frac{v_1^2}{2 c_p} 
\end{equation*}

we simply swap velocity with the Mach number. Consider that $a = \sqrt{\gamma R T_1}$. Thus, the velocity is given by $v = Ma = M \sqrt{\gamma R T_1}$. Assuming that the fluid is an ideal gas, we have know that $c_p - c_v = R$, and that $\gamma = c_p / c_v$. This leads to 

\begin{equation*}
T_{01}  = T_1 + \frac{1}{2_cp} \times M_1^2 \gamma R T_1 = T_1 + \frac{1}{2_cp} \times M_1^2 \frac{c_p}{c_v} \left(c_p - c_v \right) T_1 
\end{equation*}

This simplifies to 
\begin{equation*}
\begin{split}
T_{01} & = T_1 \left( 1 + \frac{M_1^2}{2} \left( \frac{c_p}{c_v} - 1 \right) \right) \\
T_{01} & = T_1 \left( 1 + \frac{\left( \gamma - 1 \right)}{2} M_1^2 \right) 
\end{split}
\end{equation*}

}

\end{enumerate}

### Problem 3
Air flows through a duct which has frictionless walls and constant cross-sectional area. The mass flow rate is constant at $20 \; kg/s$ and a total of ${\color[rgb]{0.334690,0.296180,0.998454}2 \; MegaWatts}$ of heat is added along the length of the tube. Measurements taken at the exit of the tube indicate that the exit static temperature is ${\color[rgb]{0.164799,0.878862,0.723179}660 \; K}$ and the exit velocity is ${\color[rgb]{0.059472,0.501943,0.998465}330 \; m/s}$. You may assume $\gamma=1.4$, ${\color[rgb]{0.986252,0.007236,0.027423}c}_{\color[rgb]{0.315209,0.728565,0.037706}p }= {\color[rgb]{0.986252,0.007236,0.027423}1005 \; J / \left( kg \cdot K \right)}$ and use the subsonic flow table provided after these questions. 

\begin{enumerate}
\item[a.] Calculate the inlet stagnation temperature (Hint: the expression you derived in 2(c) will be useful here). ${\color[rgb]{0.028509,0.250925,0.501969}[3]}$

{\color{RubineRed} 

From the provided expression we have:

\begin{equation*}
\frac{T_{02}}{T_2} = 1 + \frac{\left(\gamma -1 \right)}{2}M_2^2 
\end{equation*}

First, we work out the Mach number, $M_2$ via $a = \sqrt{\gamma R T} = \sqrt{1.4 \times 287 \times 660} = 514.964$. Thus the Mach number at the exit is $M_2=v/a = 330/515 = 0.64$. Plugging this value into the temperature expression yields

\begin{equation*}
\frac{T_{02}}{660} = 1 + \frac{\left(1.4 -1 \right)}{2}0.64^2 \Rightarrow T_{02} = 714 \; K. 
\end{equation*}

Note that for the above, students could have also used the tables. Now to work out the stagnation temperature at the inlet we use

\begin{equation*}
\dot{Q} = \dot{m} c_p \left(T_{02} - T_{01} \right) \Rightarrow T_{01} = 714 - \frac{2 \times 10^{6}}{20 \times 1005}
\end{equation*}
This leads to $T_{01} = 614.5 \; K$. 
}

\item[b.] Calculate the inlet velocity. ${\color[rgb]{0.028509,0.250925,0.501969}[4]}$

{\color{RubineRed} 
To work out the inlet velocity, we shall make use of the impulse function which remains unchanged as there is no friction. This results in 
 
\begin{equation}
\frac{ \frac{F_2}{\dot{m} \sqrt{c_p T_{02} }}}{\frac{F_1}{\dot{m} \sqrt{c_p T_{01} }}} = \sqrt{\frac{T_{01}}{T_{02}}} = \sqrt{\frac{614.5}{714}} = 0.9277
\end{equation}
 
At the exit $M_2 = 0.64$, which from the tables leads to $F_2 /\dot{m} \sqrt{c_p T_{02}} = 1.0678$. Thus, at entry $F_2 /\dot{m} \sqrt{c_p T_{01}} = 1.0678 / 0.9277 = 1.151$. The tables inform us that at this non-dimensional impulse function value, the Mach number is approximately $M_1 = 0.535$, and that the temperature ratio is $T_{1}/T_{01} = 0.946$. Thus, $T_1 = 0.946 \times 614.5 = 581.3\; K$. Using this value of the static temperature we have

\begin{equation}
a_1 = \sqrt{\gamma R T_1} = \sqrt{1.4 \times 287 \times 581.3} = 483.3 \; m/s.
\end{equation}

Using this speed of sound, the inlet velocity is $v_1 = M_1 \times a = 0.535 \times 483.3 = 258.5 \; m/s$. 
 

}

\item[c.] Draw and label a ${\color[rgb]{0.164799,0.878862,0.723179}T}-{\color[rgb]{0.500000,0.500000,0.500000}s}$ diagram to show the complete process. ${\color[rgb]{0.028509,0.250925,0.501969}[3]}$

{\color{RubineRed} 
Please see the diagram below.

\begin{figure}[h]
    \centering
    \includegraphics[width=0.4\textwidth]{images/image_P4_4.png}
\caption{Temperature vs specific entropy for the Rayleigh process.}
    \label{fig:enter-label}
\end{figure}

}

\item[d.] How much more heat would have to be added to choke the exit of the pipe (in ${\color[rgb]{0.334690,0.296180,0.998454} MegaWatts}$)? ${\color[rgb]{0.028509,0.250925,0.501969}[6]}$

{\color{RubineRed} 
To choke the exit, we would have $M_2 = 1$. At this value the non-dimensional impulse function (from the tables) is $F_2 / \dot{m} \sqrt{c_p T_{02} } = 0.9897$. 

Thus the ratio of stagnation temperatures is

\begin{equation*}
\frac{T_{02}}{T_{01}} = \left( \frac{ \frac{F_1}{\dot{m} \sqrt{c_p T_{01} }} }{\frac{F_2}{\dot{m} \sqrt{c_p T_{02} }}}  \right)^2 = \left( \frac{1.15}{0.9897} \right)^2 = 1.35
\end{equation*}

where we obtain the $1.15$ from the tables (entry corresponding to $M_1=0.535$). It is perfectly acceptable if students did not interpolate and used $1.1556$ instead. In general, I do not expect students to interpolate, and thus minor differences in solutions---provided the workflow is the same---is ok. 

Given this temperature ratio, we now have

\begin{equation*}
T_{02} = 1.35 \times T_{01} = 1.35 \times 614.5 = 892.6 \; K
\end{equation*}

This would correspond to a heat (rate) of

\begin{equation*}
\dot{Q} = 20 \times 1005 \times \left( 829.6 - 614.5\right)  = 4323510 \; W = 4.32 \; MW.
\end{equation*}

This amounts to roughly $2.16$ times more heat required to choke the exit of the pipe, or equivalently $4.32 - 2.00 = 2.32 MW$ more heat.  
}


\end{enumerate}
